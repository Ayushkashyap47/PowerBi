Ans5

```python
#dropping rows having quantity less than zero, hence just filtering out rows with quantity more than zero
df1=df[df['Quantity']>0]
```

```python
df1
```

Ans6

```python
#Unique instances of country
df1.Country.unique()
```

```python
#Country with highest instances
df1.Country.value_counts().reset_index().iloc[0]
```

Comment: United Kingdom got the highest instances/rows 

Ans7

```python
df1["Total_Sales"]=df1['Quantity']*df1['UnitPrice']
```

```python
df1
```

Ans8

```python
#Top 5 countries with highest sales
df1.groupby(by='Country').Total_Sales.sum().sort_values(ascending=False).head(5)
```

```python
#Top 5 countries with highest average sales
df1.groupby(by='Country').Total_Sales.mean().sort_values(ascending=False).head(5)
```

```python
#Top 5 products with highest sales
df1.groupby(by='StockCode').Total_Sales.sum().sort_values(ascending=False).head(5)
```
